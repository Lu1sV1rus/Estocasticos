<p><font size="6">Caminata Aleatoria 1D</font></p>

---

<p><font size="3">
Este código simula el movimiento de una partícula en una dimensión (línea recta).
<br>Es la base para entender modelos financieros como el <font color="tomato"> Movimiento Browniano</b>.
</font></p>

<hr size="2" ">

<p><font size="4">Variables y Probabilidades</font></p>

<ul>
    <p><font size="3">Probabilidad de éxito (moverse a la derecha, +1):<font color="tomato"> p</font></p>
    <p><font size="3">Probabilidad de fracaso (moverse a la izquierda, -1):<font color="tomato"> q</font></p>
    <p><font size="3">Condición:<font color="tomato"> p + q = 1</font></p>
</ul>

---

<p><font size="4" color="tomato">Librerías Utilizadas</b></font></p>

<p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<font size="3" color="tomato">import random as r</font>
<br>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<font size="2">&nbsp;&nbsp;&nbsp;&nbsp;Genera los números aleatorios para decidir el paso.</font>
</p>

<p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<font size="3" color="tomato">import matplotlib.pyplot as plt</font>
<br>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<font size="2">&nbsp;&nbsp;&nbsp;&nbsp;Se encarga de graficar el recorrido de la caminata.</font>
</p>

```python
'''Codigo de una caminata aleatoria 1D, con probabilidad p de moverse a la
derecha y probabilidad q de moverse a la izquierda, tal que p + q = 1.'''
#Importamos las liberias
import random as r
import matplotlib.pyplot as plt
```

<p><font size="5">Función: Caminata(w, c)</font></p>

<p><font size="3">
Define la lógica del movimiento paso a paso y almacena el historial de la posición.
</font></p>

<hr size="2">

<p><font size="4">Parámetros de Entrada</font></p>

<ul>
    <p><font size="3">Número total de pasos a realizar:<font color="tomato"> w</font></p>
    <p><font size="3">La probabilidad de que se mueva a la derecha (o izquierda):<font color="tomato"> c</font></p>
</ul>

---

<p><font size="4">Proceso Lógico</font></p>

<p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<font size="3" color="tomato">x = 0  y  h = []</font>
<br>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<font size="2">&nbsp;&nbsp;&nbsp;&nbsp;Se inicializa la posición en el origen y se crea la lista del historial.</font>
</p>

<p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<font size="3" color="tomato">Ciclo For (w)</font>
<br>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<font size="2">&nbsp;&nbsp;&nbsp;&nbsp;Por cada paso, se genera un número aleatorio entre 0 y 1.</font>
</p>

<p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<font size="3" color="tomato">If (p > c)</font>
<br>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<font size="2">&nbsp;&nbsp;&nbsp;&nbsp;Si el número es mayor al umbral, avanza (+1); de lo contrario, retrocede (-1).</font>
</p>

---

<p><font size="4">Salida del Código</font></p>

<ul>
    <p><font size="3">Imprime la posición final de la partícula:<font color="tomato"> print(x)</font></p>
    <p><font size="3">Devuelve la lista con todos los puntos visitados:<font color="tomato"> return h</font></p>
</ul>

```python
def caminata(w,c):
  x=0
  h=[]
  h.append(x)
  for i in range(w):
    p=r.random()
    if p>c:
      x+=1
      h.append(x)
    else:
      x-=1
      h.append(x)
  print(x)
  return h
```

<p><font size="5">Ejecución del Modelo</font></p>

<p><font size="3">
Se definen las condiciones iniciales y se llama a la función para obtener los resultados de la simulación.
</font></p>

<hr size="2">

<p><font size="4">Configuración de Variables</font></p>

<ul>
    <p><font size="3">Probabilidad límite (50% para cada lado):<font color="tomato"> c = 0.5</font></p>
    <p><font size="3">Cantidad de iteraciones en la serie:<font color="tomato"> w = 100</font></p>
</ul>

---

<p><font size="4">Llamada a la Función</font></p>

<p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<font size="3" color="tomato">s1 = caminata(w, c)</font>
<br>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<font size="2">&nbsp;&nbsp;&nbsp;&nbsp;Se ejecuta el proceso y se almacena la lista de posiciones en la variable "s1".</font>
</p>

---

<p><font size="4">Interpretación Financiera</font></p>

<p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<font size="2">Al usar un <font color="tomato">c = 0.5</font>, es un camino aleatorio con la misma oportunidad de crecer como decrecer</font>
</p>

```python
# Probabilidad
c=0.5
# Pasos
w=100
# Funcion
s1=caminata(w,c)
```

<p><font size="5">Función: ceros(s1)</font></p>

<p><font size="3">
Contabiliza cuántas veces la caminata aleatoria regresa o cruza el punto de origen (cero).
</font></p>

<hr size="2">

<p><font size="4">Lógica del Contador</font></p>

<p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<font size="3" color="tomato">g = -1</font>
<br>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<font size="2">&nbsp;&nbsp;&nbsp;&nbsp;Se inicializa en -1 para no contar la posición inicial (el cero de arranque).</font>
</p>

<p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<font size="3" color="tomato">for i in range(len(s1))</font>
<br>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<font size="2">&nbsp;&nbsp;&nbsp;&nbsp;Recorre toda la lista de posiciones generada en la simulación.</font>
</p>

---

<p><font size="4">Condición</font></p>

<ul>
    <p><font size="3">Si la posición actual es igual a cero:<font color="tomato"> if s1[i] == 0</font></p>
    <p><font size="3">Aumenta el contador de encuentros:<font color="tomato"> g += 1</font></p>
</ul>

---

<p><font size="4">Resultado Final</font></p>

<p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<font size="3" color="tomato">return(g)</font>
<br>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<font size="2">&nbsp;&nbsp;&nbsp;&nbsp;Devuelve el número total de veces que paso por el origen.</font>
</p>

```python
# Extra conteos de ceros
def ceros(s1):
  g=-1
  for i in range(len(s1)):
    if s1[i]==0:
      g+=1
  return(g)

ceros(s1)
```

<p><font size="3">
La probabilidad <font color="tomato"> c = 0.2</font> favorece el crecimiento.
</font></p>

<hr size="2">

<p><font size="4">Configuración del Drift</font></p>

<ul>
    <p><font size="3">Nueva probabilidad límite:<font color="tomato"> c = 0.2</font></p>
    <p><font size="3">Interpretación:<font color="tomato"> 80% derecha / 20% izquierda</font></p>
</ul>

---

<p><font size="4">Ejecución y Gráfica</font></p>

<p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<font size="3" color="tomato">s2 = caminata(w, 0.2)</font>
<br>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<font size="2">&nbsp;&nbsp;&nbsp;&nbsp;Al ser el umbral muy bajo, la condición "p > c" se cumple la mayoría de las veces.</font>
</p>

<p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<font size="3" color="tomato">plt.plot(s2)</font>
<br>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<font size="2">&nbsp;&nbsp;&nbsp;&nbsp;Grafica la serie de datos.</font>
</p>

```python
# Funcion con tendencia a la derecha (crece)
# Funcion s2
s2=caminata(w,0.2)
# Grafica
plt.figure(figsize=(12,4))
plt.axhline(y=0, linestyle="--", color='red')
plt.title('Caminata tendencia a la derecha 1D')
plt.xlabel('Iteraciones')
plt.ylabel('Valor')
plt.plot(s2, color='green')
```

<p><font size="5">S3: Simulación con tendencia negativa</font></p>

<p><font size="3">
La probabilidad <font color="tomato"> c = 0.8</font> favorece el decrecimiento.
</font></p>

<hr size="2">

<p><font size="4">Configuración del Drift</font></p>

<ul>
    <p><font size="3">Nueva probabilidad límite:<font color="tomato"> c = 0.8</font></p>
    <p><font size="3">Interpretación:<font color="tomato"> 20% derecha / 80% izquierda</font></p>
</ul>

---

<p><font size="4">Ejecución y Gráfica</font></p>

<p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;

</p>

<p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<font size="3" color="tomato">plt.plot(s3)</font>

```python
# Funcion con tendencia a la izquierda (decrece)
# Funcion s3
s3=caminata(w,0.8)
# Grafica
plt.figure(figsize=(12,4))
plt.axhline(y=0, linestyle="--", color='red')
plt.title('Caminata con tendencia a la izquierda 1D')
plt.xlabel('Iteraciones')
plt.ylabel('Valor')
plt.plot(s3, color='orange')
```
