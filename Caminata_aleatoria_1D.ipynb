```python
# Codigo 1D, aleatorio, p+q=1
import random as r
#import random.randint as r
```

```python
def caminata(w,c):
  x=0
  h=[]
  h.append(x)
  for i in range(w):
    p=r.random()
    if p>c:
      x+=1
      h.append(x)
    else:
      x-=1
      h.append(x)
  print(x)
  return h
```

```python
# Probabilidad
c=0.5
# Pasos
w=100
# Funcion
s1=caminata(w,c)
```

```python
g=0
for i in range(len(s1)):
  if s1[i]==0:
    g+=1
print(g)
```

```python
from matplotlib import lines
# Grafica
import matplotlib.pyplot as plt
plt.figure(figsize=(12,4))
plt.axhline(y=0, linestyle="--", color='red')
plt.title('Caminata aleatoria 1D')
plt.xlabel('Valor')
plt.ylabel('Iteraciones')
plt.plot(s1)
```
